```python
import numpy as np
import matplotlib.pyplot as plt
L=1.0
A=1.0
B=-1.0
c=343.0
f=340.0
k=2*np.pi*f/c
x_np=np.linspace(0,1,500)
p_true = (A * np.cos(k*x_np)
              + ((B - A*np.cos(k*L)) / np.sin(k*L))*np.sin(k*x_np))
plt.figure(figsize=(10,6))
plt.plot(x_np, p_true,  label='Analytical', linewidth=2)
#plt.plot(x, p_pred, '--', label='PINN (L-BFGS)', linewidth=2)
plt.title(f'Acoustic Pressure at f = {f} Hz (k = {k:.2f} rad/m)')
plt.xlabel('x'); plt.ylabel('p(x)')
plt.grid(which='both', linestyle=':')
#plt.legend(); plt.tight_layout()
plt.show()
```
